## Correlation with heat map
### A correlation heatmap uses coloured cells, typically in a monochromatic scale, to show a 2D correlation matrix between two discreate dimentions or event types. It is very important in feature selection.

```python
df.corr() 
#it is showing the amount of correlation between different entities like totalbill and tip have 67% correlation
```

```python
sns.heatmap(df.corr())       
#the different colors show the amount of correlation based on the correlation matrix that we made
```

### Univariate analysis is the simplest form of data analysis where the data being analyzed contains only one variable. Since it's a single variable it doesn't deal with causes or relationships. The main purpose of univariate analysis is to describe the data and find patterns that exist within it.

# Joint plot
### A joint plot allows to study the relationship between 2 numeric variables. the central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

### Join plot helps to do UNIVARIATE and BIVARIATE analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#this is bivariate analysis
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
#this is bvariate analysis
```

# Pair plot
## A pairplot is also known as scatter plot, in which one variable in the same data row id matched with another variable's value, like this: Pairs plots are just elaborations on this showing all variables period with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
#hue is used for classification
```

# Count plot
## In count plot we have to give either the x value or the y value

```python
sns.countplot(x='sex',data=df)   #for the x axis result
```

```python
sns.countplot(x='day',data=df)
```

```python
sns.countplot(y='sex',data=df)    #for y axis result
```

# Bar plot
## In bar plot we have to give both the x and y value

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='day',y='tip',data=df)
```

# Box plot

```python
sns.boxplot(x='sex',y='tip',data=df)
```

# Violin plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
