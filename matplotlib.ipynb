```python
import matplotlib.pyplot as plt
%matplotlib inline      
#as we have written this line we do not have to write plot.show every time to see the output
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

# Scatter Plot

```python
## Scatter plt

plt.scatter(x,y,c='g')
```

```python
plt.scatter(x,y,c='g')    # the value of c gives the color
plt.xlabel('X axis')    # it labels the x axis
plt.ylabel('Y axis')    # it labels the y axis
plt.title('Graph in 2D')    # it gives title to the graph
```

# Line Plot

```python
y=x*x
```

```python
## Plot plt
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Example of plot graph')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,'r--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('sine wave form')
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set the subplot grid that has height 2 and width 2

plt.subplot(2,2,1)
plt.plot(x,y_sin,'r*-')
plt.title('Sin')

#set the subplot grid that has height 2 and width 2

plt.subplot(2,2,2)
plt.plot(x,y_cos,'b*-')
plt.title('Cos')
```

# Bar Graph

```python
#Bar Plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Histogram

```python
a=np.array([10,20,30,40,50,60,70,80,90,14,25,36])
plt.hist(a,color='r')
plt.title('Histogram')
plt.show()    #this is used to show the graph,it is not necessary here but it is required in IDE like spyder
```

# Pie Chart

```python
#piechart

labels=['Python','JAVA','Ruby','Scala']
sizes=[212,256,145,301]
colors=['yellow','red','blue','green']
explode=[0.1,0,0,0]  #explode makes any slice to move outwards

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
